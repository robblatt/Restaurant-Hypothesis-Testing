```python
import requests
```

```python
r1 = requests.get(
    'https://data.cityofnewyork.us/resource/fhrw-4uyv.json?complaint_type=Rodent&$where="created_date" > "2017-06-05T00:00:00.000"&$offset=0')
response1 = r1.json()
df1_main = pd.DataFrame(response1)
```

```python
offset_ = 0
while offset_ < 2000000:
    df1_a = df1_main
    r = requests.get('https://data.cityofnewyork.us/resource/fhrw-4uyv.json?complaint_type=Rodent&$where="created_date" > "2017-06-05T00:00:00.000"&$offset=' + str(offset_))
    response = r.json() 
    df1_b = pd.DataFrame(response)
    df1_main = pd.concat([df1_a, df1_b], sort=True)
    offset_ += 1000
    if offset_ % 10000 == 0:
        print(offset_,' ', len(df1_main))
        
```

```python
df1_main.reset_index(inplace=True)
```

```python
df1_main = df1_main.drop_duplicates(subset=['created_date', 'incident_address'], keep='first')
```

```python
df1_main
```

```python
df1_main.to_csv('df_rat_complaints_complete.csv')
```
